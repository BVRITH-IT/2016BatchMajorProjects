```python
def https_token(l):
        l=str(l)
        match=re.search('https://|http://',l)
        try:
            if match.start(0)==0 and match.start(0) is not None:
                l=l[match.end(0):]
                match=re.search('http|https',l)
                if match:
                    return 1
                else:
                    return 0
        except:
            return 1
```
